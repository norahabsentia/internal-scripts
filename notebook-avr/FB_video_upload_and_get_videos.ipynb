#### Script to upload videos

```python
def upload_videos():
    found_count = 0
    uploaded_count = 0
    app_id = APP_ID
    app_secret = APP_SECRET
    access_token = ACCESS_TOKEN
    FacebookAdsApi.init(app_id, app_secret, access_token)
    account = AdAccount("act_%s" % AD_ACCOUNT)
    existing_videos = list(account.get_ad_videos(fields=["title"]))
    ad_video = AdVideo(parent_id="act_%s" % AD_ACCOUNT)
    for index, name in enumerate(os.listdir(BASEPATH)):
        found = False
        print(name)
        for video in existing_videos:
            title = video.get("title", "")
            if name == title:
                found = True
        if found:
            found_count += 1
            print("%d-video %s already present skipping - Total count %d" % (index, name, found_count))
            continue
        uploaded_count += 1
        print("%d-video %s not found. Add it" % (index, name))
        path = "%s%s" % (BASEPATH, name)
        ad_video[AdVideo.Field.filepath] = path
        response = ad_video.remote_create()
        print(response)
    print("Videos already present = %s" % found_count)


upload_videos()
```
